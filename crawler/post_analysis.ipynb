```python
# Standard library imports
import os
import sys
from datetime import datetime
import glob

# Third-party imports
import pandas as pd
import numpy as np
import jieba
from ipywidgets import widgets
from IPython.display import display, clear_output
```

```python
print(f"Python version: {sys.version}")
```

```python
latest_file = max(glob.glob("*_flow_items.csv"))

# Load the file into a pandas dataframe
df = pd.read_csv(latest_file)
df.head(10)
```

```python
# user posts rank by user nickname
top_users = df['user_nickname'].value_counts().reset_index()
top_users.columns = ['User', 'Count']
top_users = top_users.head(10)
top_users.index = top_users.index + 1  # Start ranking from 1 instead of 0

print("Top 10 Users by Number of Entries:")
print("----------------------------------")
top_users
```

```python
# user posts rank by user_hash_id/user_view_url, since some of the user's nickname will be "已注销" if they terminate their account.
top_users = df['user_view_url'].value_counts().reset_index()
top_users.columns = ['User', 'Count']
top_users = top_users.head(10)
top_users.index = top_users.index + 1  # Start ranking from 1 instead of 0

print("Top 10 Users by Number of Entries:")
print("----------------------------------")
top_users
```

```python
# title keyword search and filter by blacklist
def search_titles(keywords_text, blacklist_text):
    # Split the inputs into lists, strip whitespace, and remove empty strings
    keywords = [k.strip() for k in keywords_text.split(',') if k.strip()]
    blacklist = [k.strip() for k in blacklist_text.split(',') if k.strip()]
    
    if not keywords:
        print("Please enter at least one keyword")
        return
        
    pattern = '|'.join(keywords)
    search_results = df[df['title'].str.contains(pattern, case=False, na=False)]
    
    if blacklist:
        blacklist_pattern = '|'.join(blacklist)
        search_results = search_results[~search_results['title'].str.contains(blacklist_pattern, case=False, na=False)]
    
    results_table = search_results[['title', 'user_nickname', 'view_url']]
    
    clear_output(wait=True)
    print(f"Found {len(results_table)} items containing any of these keywords: {keywords}")
    if blacklist:
        print(f"(Excluding items containing: {blacklist})")
    print("----------------------------------")
    
    # Set pandas display options for fuller view
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', None)
    
    # Create a scrollable output widget
    output = widgets.Output(layout={'border': '1px solid black', 
                                  'max_height': '600px', 
                                  'overflow': 'auto',
                                  'width': '100%'})
    with output:
        display(results_table)
    display(output)

# Create search box widgets
search_box = widgets.Text(
    value='',
    placeholder='Enter keywords separated by commas',
    description='Search:',
    disabled=False,
    layout={'width': '50%'}
)

blacklist_box = widgets.Text(
    value='',
    placeholder='Enter blacklist keywords separated by commas',
    description='Exclude:',
    disabled=False,
    layout={'width': '50%'}
)

# Create and display the interactive widget
widgets.interactive(search_titles, 
                   keywords_text=search_box,
                   blacklist_text=blacklist_box)
```

```python
# parse chinese sentences and save
current_date = datetime.now().strftime('%Y%m%d')
output_filename = f"{current_date}_flow_items_title_cut.csv"

# Check if file already exists
if not os.path.exists(output_filename):
    # Create a new column with the segmented titles
    df['title_cut'] = df['title'].apply(lambda x: ' '.join(jieba.cut(str(x))))
    
    # Save to new CSV file with date
    df.to_csv(output_filename, index=False)
    print(f"Saved processed data to {output_filename}")
else:
    print(f"File {output_filename} already exists, skipping processing")
```

```python
# high frequency word rank
filtered_words = [
    '什么', '一个', '这个', '那个', '就是', '可以', '没有', '不是', '如何', '为什么',
    '怎么', '现在', '知道', '觉得', '需要', '这样', '那样', '自己', '因为', '所以',
    '还是', '不要', '真的', '已经', '可能', '应该', '时候', '出来', '看看', '一下',
    '一直', '很多', '问题', '不能', '感觉', '时间', '东西', '事情', '发现', '一些',
    '大家', '然后', '直接', '买家', '但是', '这么', '你们', '开始', '这种',
    '看到', '最后', '关于', '还有', '自后', '如果', '是不是', '确实', '只是', '那么',
    '智能', '各位', '真是', '而且', '后面', '几个', '他们', '10', '还要', '并且'  # Added new words
]

# Get the latest title_cut file and process data
latest_cut_file = max(glob.glob("*_flow_items_title_cut.csv"))
df_cut = pd.read_csv(latest_cut_file)
all_words = ' '.join(df_cut['title_cut'].fillna('')).split()

# Filter words: remove single characters and filtered words
all_words = [word for word in all_words if len(word) > 1 and word not in filtered_words]

# Create word frequency dataframe
word_freq = pd.Series(all_words).value_counts().reset_index()
word_freq.columns = ['Word', 'Frequency']
word_freq.index = word_freq.index + 1

# Set pandas display options for fuller view
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print("Most Frequent Words (excluding single characters and common words):")
print("----------------------------------")

# Create a scrollable output widget
output = widgets.Output(layout={'border': '1px solid black', 
                              'max_height': '600px', 
                              'overflow': 'auto',
                              'width': '100%'})
with output:
    display(word_freq)
display(output)
```
