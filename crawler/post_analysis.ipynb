```python
latest_file = max(glob.glob("*_flow_items.csv"))

# Load the file into a pandas dataframe
df = pd.read_csv(latest_file)
df.head(10)
```

```python
# title keyword search and filter by blacklist
def search_titles(keywords_text, blacklist_text):
    # Split the inputs into lists, strip whitespace, and remove empty strings
    keywords = [k.strip() for k in keywords_text.split(',') if k.strip()]
    blacklist = [k.strip() for k in blacklist_text.split(',') if k.strip()]
    
    if not keywords:
        print("Please enter at least one keyword")
        return
        
    pattern = '|'.join(keywords)
    search_results = df[df['title'].str.contains(pattern, case=False, na=False)]
    
    if blacklist:
        blacklist_pattern = '|'.join(blacklist)
        search_results = search_results[~search_results['title'].str.contains(blacklist_pattern, case=False, na=False)]
    
    results_table = search_results[['title', 'user_nickname', 'view_url']]
    
    clear_output(wait=True)
    print(f"Found {len(results_table)} items containing any of these keywords: {keywords}")
    if blacklist:
        print(f"(Excluding items containing: {blacklist})")
    print("----------------------------------")
    
    # Set pandas display options for fuller view
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', None)
    
    # Create a scrollable output widget
    output = widgets.Output(layout={'border': '1px solid black', 
                                  'max_height': '600px', 
                                  'overflow': 'auto',
                                  'width': '100%'})
    with output:
        display(results_table)
    display(output)

# Create search box widgets
search_box = widgets.Text(
    value='',
    placeholder='Enter keywords separated by commas',
    description='Search:',
    disabled=False,
    layout={'width': '50%'}
)

blacklist_box = widgets.Text(
    value='',
    placeholder='Enter blacklist keywords separated by commas',
    description='Exclude:',
    disabled=False,
    layout={'width': '50%'}
)

# Create and display the interactive widget
widgets.interactive(search_titles, 
                   keywords_text=search_box,
                   blacklist_text=blacklist_box)
```

```python
# parse chinese sentences and save
current_date = datetime.now().strftime('%Y%m%d')
output_filename = f"{current_date}_flow_items_title_cut.csv"

# Check if file already exists
if not os.path.exists(output_filename):
    # Create a new column with the segmented titles
    df['title_cut'] = df['title'].apply(lambda x: ' '.join(jieba.cut(str(x))))
    
    # Save to new CSV file with date
    df.to_csv(output_filename, index=False)
    print(f"Saved processed data to {output_filename}")
else:
    print(f"File {output_filename} already exists, skipping processing")
```
